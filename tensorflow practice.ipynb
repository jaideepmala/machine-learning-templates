```python
#batch norm implementation
from tensorflow.contrib.layers import batch_norm
is_training_batch = tf.placeholder(tf.bool, shape=(), name='is_training_batch')
bn_params = {
'is_training': is_training_batch,
'decay': 0.99,
'updates_collections': None
}
```

```python
# Drop out implementation
from tensorflow.contrib.layers import dropout
is_training_drop = tf.placeholder(tf.bool,shape=(),name="is_training_drop")
keep_prob=0.8
X_drop = dropout(X,keep_prob,is_training = is_training_drop)
```

```python
#drop out implementation
# with tf.name_scope("DNN"):
#     hlayer1 = neuron_layer(X_drop,hl_1,"hiddenlayer1",distribution="normal",activation="elu")
#     hlayer1_drop = dropout(hlayer1,keep_prob,is_training=is_training)
#     hlayer2 = neuron_layer(hlayer1_drop,hl_2,"hiddenlayer2",distribution = "normal",activation="elu")
#     hlayer2_drop = dropout(hlayer2,keep_prob,is_training=is_training)
#     output = neuron_layer(hlayer2_drop,ol,"outputlayer")
    
```

```python
from tensorflow.contrib.layers import fully_connected
with tf.name_scope("MODEL"):
    with tf.contrib.framework.arg_scope([fully_connected],normalizer_fn=batch_norm,normalizer_params=bn_params):
        hlayer1 = fully_connected(X_drop,hl_1,activation_fn=tf.nn.elu,weights_initializer=tf.contrib.layers.variance_scaling_initializer())
        hlayer1_drop = dropout(hlayer1,keep_prob,is_training=is_training_drop)
        hlayer2 = fully_connected(hlayer1_drop,hl_2,activation_fn=tf.nn.elu,weights_initializer=tf.contrib.layers.variance_scaling_initializer())
        hlayer2_drop = dropout(hlayer2,keep_prob,is_training=is_training_drop)
        output = fully_connected(hlayer2_drop,ol)
```

```python
with tf.name_scope("LOSS"):
    entropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,logits=output)
    loss = tf.reduce_mean(entropy,name ="loss")
```

```python
learning_rate = 0.01
with tf.name_scope("TRAIN"):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
```

```python
with tf.name_scope("EVAL"):
    y_pred = tf.cast(tf.argmax(output, axis=1), tf.int64)
    matched = tf.equal(y_pred, y)
    accuracy = tf.reduce_mean(tf.cast(matched, tf.float32))

#     correct = tf.nn.in_top_k(output, y, 1)
#     accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```
